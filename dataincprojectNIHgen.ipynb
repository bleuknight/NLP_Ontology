```python
##get everything into dataframes for cross referencing
#PRJ files give the application number (needed for project abstracts)
import pandas as pd
def readMyProj(filename):
    #imported as object to save memory, will convert to appropriate data types as needed
    df = pd.read_csv(filename, encoding='cp1252', dtype = object) 
    # note not utf8 like other files
    #details for analysis : total cost -how much does a publication cost? STUDY_SECTION_NAME who is giving out the most $
    #for fun Org_city and Org_name- where does the money go? Application type (#1 is new) for time to pub
    keepColumns = ['APPLICATION_ID', 'APPLICATION_TYPE', 'CORE_PROJECT_NUM', 'FULL_PROJECT_NUM', 'ORG_NAME',
                'ORG_STATE', 'STUDY_SECTION_NAME', 'DIRECT_COST_AMT','INDIRECT_COST_AMT','TOTAL_COST']
    #new truncated DataFrame
    dfProjDetails = df[keepColumns]
    return dfProjDetails

dfProjDetails = readMyProj('RePORTER_PRJ_C_FY2013.csv')
dfProjDetails.head()
```

```python
def getKeys(dfProjDetails):
    keys = ['APPLICATION_ID', 'CORE_PROJECT_NUM', 'FULL_PROJECT_NUM']
    dfkeys = dfProjDetails[keys]
    #set index to merge with other files
    dfkeys.set_index('APPLICATION_ID', inplace = True)
    return dfkeys

dfkeys = getKeys(dfProjDetails)
dfkeys.head()

    


```

```python
def readProjAbs(filename):
    #this gives the application ID and the abstract text- 
    #do the publications overlap with the abstract text? to what degree?
    prjAbs = pd.read_csv(filename, encoding='cp1252', dtype = object)
    prjAbs.set_index('APPLICATION_ID', inplace = True)
    GrantAbs = pd.concat([dfkeys, prjAbs], axis=0, join='outer')
    print (GrantAbs.head())
    return GrantAbs

GrantAbs = readProjAbs('RePORTER_PRJABS_C_FY2013.csv')
    
```

```python
def links(filename):
    #this is the file which links 2016 publications to grant numbers
    link = pd.read_csv(filename)
    pmids = list(link['PMID'])
    #there are ~2000 publications- I need to split them up in case my connection is interrupted
    split = [pmids[i:i+500] for i in range(0,len(pmids),500)]
    return split
split = links("2013links.csv")
LL = list(range(0, len(split)))
print(LL)
```

```python
#import sleep to pause between queries
from time import sleep
import json
import pandas as pd
def query(split, LL):  
    #using the biopython package to interface with pubmed api
    from Bio import Entrez
    master_dict = []
    for n in LL:
        #creates open dictionary to store abstract text with publication number
        abstract_dict = {}
        #creates list to store entries without abstract text
        without_abstract = []
        #identifier for biopython
        Entrez.email = 'scientistbleu@gmail.com'
        handle = Entrez.efetch(db="pubmed", id=','.join(map(str, split[n])),
                               rettype="xml", retmode="text")
        records = Entrez.read(handle)
        for pubmed_article in records['PubmedArticle']:
            pmid = int(str(pubmed_article['MedlineCitation']['PMID']))
            article = pubmed_article['MedlineCitation']['Article']
            if 'Abstract' in article:
                abstract = article['Abstract']['AbstractText'][0]
                abstract_dict[pmid] = abstract
            else:
                without_abstract.append(pmid)
            sleep = (10)
        with open('abstract_dict.txt', 'w') as outfile:  
            json.dump(abstract_dict, outfile)
        with open('without_abstract.txt', 'w') as outfile:  
            json.dump(without_abstract, outfile)
        master_dict.append(abstract_dict.copy())
    return master_dict
master_dict = query(split, LL)
with open('master_dict.txt', 'w') as outfile:  
    json.dump(master_dict, outfile)




```
